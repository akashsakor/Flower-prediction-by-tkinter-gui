```python
from sklearn.datasets import load_iris
df=load_iris()
import pandas as pd
import numpy as np

x=df.data
y=df.target
from sklearn.model_selection import train_test_split
xtrain,xtest,ytrain,ytest=train_test_split(x,y,train_size=0.75,random_state=1)
from sklearn.neighbors import KNeighborsClassifier
model=KNeighborsClassifier(n_neighbors=3)
model.fit(xtrain,ytrain)
ypred=model.predict(xtest)

model.score(xtest,ytest)
from sklearn.metrics import confusion_matrix

from sklearn.metrics import accuracy_score

from sklearn.metrics import classification_report

ypred=model.predict(xtest)

cm=confusion_matrix(ytest,ypred) 

###########################################################################

import tkinter
from tkinter import *
#window
win = Tk()  
win.geometry('400x150')  
win.title('Tkinter')

a='setosa'
b='versicolor'
c='virginica'

k=IntVar()
p=IntVar()
m=IntVar()
n=IntVar()
ent1=Entry(win,textvariable=k).pack()
ent2=Entry(win,textvariable=p).pack()
ent3=Entry(win,textvariable=m).pack()
ent4=Entry(win,textvariable=n).pack()

def say_hi(event):
    o=model.predict([[k.get(),p.get(),m.get(),n.get()]])
    print(o)
    if o==0: 
        pp = Label(win, text=a, font="BOLD")
        pp.pack()
    elif o==1:        
        pp = Label(win, text=b, font="BOLD")
        pp.pack()
    else:        
        pp = Label(win, text=c, font="BOLD")
        pp.pack()

btn=tkinter.Button(win, text="PREDICT")
btn.bind("<Button-1>",say_hi)

btn.pack()

def remove_text():
    pp.config(text="")
    
def func():
    win.destroy()
    

def clear():
    global exp
    exp=""
    k.set(exp)
    p.set(exp)
    m.set(exp)
    n.set(exp)    

l1=Label(win,text='sepal length (cm)').place(x=30)
l2=Label(win,text='sepal width (cm)').place(x=30,y=20)
l3=Label(win,text='petal length (cm)').place(x=30,y=40)
l4=Label(win,text='petal width (cm)').place(x=30,y=60)


but1=Button(win,text="clear",command=clear).pack()
btn=Button(win,text='close',command=func).pack()

win.mainloop()
```
